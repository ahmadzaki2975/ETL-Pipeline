```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Memuat data dari file CSV
file_path = 'ready-merge.csv'
data = pd.read_csv(file_path)

data.head()
```

# Pola Perubahan Cuaca dan Kualitas Udara dari Waktu ke Waktu:

```python
import matplotlib.pyplot as plt

# Ubah kolom jamCuaca menjadi format waktu jika belum
data['jamCuaca'] = pd.to_datetime(data['jamCuaca'])

# Pisahkan data menjadi series waktu untuk tempC dan iqa
time_series_tempC = data.set_index('jamCuaca')['tempC']
time_series_iqa = data.set_index('jamCuaca')['iqa']

# Plot time series untuk tempC
plt.figure(figsize=(10, 6))
plt.plot(time_series_tempC.index, time_series_tempC, label='tempC', color='blue')
plt.title('Grafik Time Series Variabel tempC')
plt.xlabel('Waktu')
plt.ylabel('tempC')
plt.legend()
plt.grid(True)
plt.show()

# Plot time series untuk iqa
plt.figure(figsize=(10, 6))
plt.plot(time_series_iqa.index, time_series_iqa, label='IQA', color='red')
plt.title('Grafik Time Series Variabel IQA')
plt.xlabel('Waktu')
plt.ylabel('IQA')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf

# Plot autokorelasi untuk tempC
plot_acf(time_series_tempC, lags=30)  # Sesuaikan nilai lags sesuai kebutuhan
plt.title('Autokorelasi Variabel tempC')
plt.xlabel('Lag')
plt.ylabel('Autokorelasi')
plt.show()

# Plot autokorelasi untuk iqa
plot_acf(time_series_iqa, lags=30)  # Sesuaikan nilai lags sesuai kebutuhan
plt.title('Autokorelasi Variabel IQA')
plt.xlabel('Lag')
plt.ylabel('Autokorelasi')
plt.show()
```
