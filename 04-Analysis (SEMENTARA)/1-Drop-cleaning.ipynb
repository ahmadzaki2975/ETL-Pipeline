```python
import pandas as pd

# Memuat data dari file CSV
file_path = '/Users/erikuncoro/Documents/Project_Rekdat/ETL-Pipeline/04-Analysis/four-days-merge.csv'
data = pd.read_csv(file_path)

# Melakukan drop kolom 'id', 'cuaca', 'accessed'
data = data.drop(['id', 'cuaca', 'accessed'], axis=1)

# Menyimpan data yang telah dimodifikasi ke dalam file CSV baru
new_file_path = 'ready-merge.csv'
data.to_csv(new_file_path, index=False)

print(f"Data has been saved to {new_file_path}")
```
